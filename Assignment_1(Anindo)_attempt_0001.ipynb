```python
# Step 19 Visualization of Data
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame creation
data = pd.DataFrame({
    'LocationName': ['Location1', 'Location2', 'Location3', 'Location4'],
    'ISO': ['ISO1', 'ISO2', 'ISO3', 'ISO4'],
    'country': ['Country1', 'Country2', 'Country1', 'Country3'],
    'region': ['Region1', 'Region2', 'Region1', 'Region2'],
    'world_region': ['World1', 'World2', 'World1', 'World2'],
    'MPI': [0.1, 0.2, 0.1, 0.3],
    'geo': ['Geo1', 'Geo2', 'Geo1', 'Geo2'],
    'lat': [10.0, 20.0, 10.0, 20.0],
    'lon': [30.0, 40.0, 30.0, 40.0]
})

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a histogram for the 'region' column
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.countplot(data=data, x='region', order=data['region'].value_counts().index)
plt.title('Distribution of Regions')
plt.xlabel('Region')
plt.ylabel('Count')

# Create a histogram for the 'country' column
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.countplot(data=data, x='country', order=data['country'].value_counts().index)
plt.title('Distribution of Countries')
plt.xlabel('Country')
plt.ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```
